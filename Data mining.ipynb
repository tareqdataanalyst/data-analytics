```python
import pandas as pd
taxi_data = pd.read_csv('C2_2017_Yellow_Taxi_Trip_Data.csv')
```

```python
taxi_data.head(10)
```

```python
taxi_data.info()  
                      #No null values, 
                #int, float & object dtypes. dtypes are non numeric two of which are datetime.
                #1st column is not labelled.
```

```python
taxi_data.describe()
                    #Regarding trip distance, most rides are between 1-3 miles, but the maximum is over 33 miles. 
                    #Regarding fare amount, the distribution is worth considering. 
                    #The maximum fare amount is a much larger value ($1000) than the 25-75 percent range of values. 
                    #Also, it's questionable how there are negative values for fare amount.
```

```python

# Sort the data by trip distance from maximum to minimum
taxi_data_sorted = taxi_data.sort_values(by='trip_distance',ascending = False)
taxi_data_sorted
```

```python
# Sort the data by total amount and print the top 20 values
taxi_data_sorted2 = taxi_data.sort_values(by = 'total_amount',ascending = False)[['total_amount']]
taxi_data_sorted2.head(20)
```

```python
taxi_data_sorted2.tail(20)

#The values align with our earlier data discovery, where we noticed that the longest rides are approximately 33 miles.
# Yes, the first two values are significantly higher than the others.
#The most expensive rides are not necessarily the longest ones.
```

```python
# How many of each payment type are represented in data ?
taxi_data['payment_type'].value_counts()
```

```python
# What is the average tip for trips paid for with credit card?
avg_tip_cc = taxi_data[taxi_data['payment_type'] == 1]['tip_amount'].mean()
print(avg_tip_cc)

# What is the average tip for trips paid for with cash?
avg_tip_cash = taxi_data[taxi_data['payment_type'] == 2]['tip_amount'].mean()
print(avg_tip_cash)
```

```python
# How many times is each vendor ID represented in the data?
taxi_data['VendorID'].value_counts()
```

```python
# What is the mean total amount for each vendor?
taxi_data.groupby(['VendorID']).mean(numeric_only=True)[['total_amount']]
```

```python
# Filter the data for credit card payments only
credit_card = taxi_data[taxi_data['payment_type'] == 1]

# Filter the data for passenger count only
credit_card['passenger_count'].value_counts()
```

```python
# Calculate the average tip amount for each passenger count (credit card payments only)
credit_card.groupby(['passenger_count']).mean(numeric_only=True)['tip_amount']
```
